```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RepeatedStratifiedKFold,GridSearchCV
from sklearn.metrics import accuracy_score
import numpy as np
from collections import Counter
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import cycle
```

```python
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
#Check null value
print(f'% of nan before concat is {df_train.loc[df_train.isna().any(axis=1)].shape[0]/df_train.shape[0]*100:.2f}')
#Fill null value by mean
df_train=pd.concat([df_train.iloc[:,0:-1].fillna(df_train.iloc[:,0:-1].mean()),df_train.iloc[:,[-1]]],axis=1)
```

```python
#Check Imbalance Label
df_train.iloc[:,[-1]].value_counts()
```

```python
#Tranform pandas to array
y=df_train.iloc[:,[-1]].values.ravel()
X=df_train.iloc[:,0:-1].values

#Split data to make model and validation
X, X_vali, y, y_vali = train_test_split(X, y, train_size=0.7, shuffle=True, stratify=y, random_state=547)

print(f'Label on make train model 0 :  {Counter(y)[0]}  1: {Counter(y)[1]}')
print(f'Label on validation model 0 : {Counter(y_vali)[0]}  1: {Counter(y_vali)[1]}') #unseen data
```

```python
#Histrogram of each attribute
index = [0, 1, 2, 3, 4]
col_idx = list(zip(cycle(index), df_train.iloc[:, 0:-1].columns))
row=0
fig,ax=plt.subplots(nrows=5,ncols=4,figsize=(18, 10))

for i,j in col_idx:
    sns.histplot(ax=ax[i,row],data=df_train.iloc[:,0:-1],x=j)
    if i==4:
        row+=1
```

```python
#Most attribute histograms are similar to the normal distribution, the data is converted by StandardScaler

stan_d = StandardScaler()

#Use under sampling to balance label
undersample = RandomUnderSampler()
underX, under_y = undersample.fit_resample(X, y)
stan_underX=stan_d.fit_transform(underX)
```

```python
f_model = RandomForestClassifier()
f_model.fit(underX, under_y)
f_importance = f_model.feature_importances_
```

```python
#Choose  variables that have impurity greater than 0.03
df_im=pd.DataFrame([[x,f_importance[x] ]for x in range(len(f_importance))]).rename(columns={0:'attribute',1:'importance'})
fea_col=['x'+str(x) for x in df_im.loc[df_im['importance']>=0.03].attribute]
df_im.loc[df_im['importance']>=0.03]
```

```python
# remove attributes that have impurity less than 0.03 in train and validation data
train_=pd.DataFrame(underX)
X_train = train_[df_im.loc[df_im['importance']>=0.03].attribute].values

vali_ = pd.DataFrame(X_vali)
X_vali = vali_[df_im.loc[df_im['importance']>=0.03].attribute].values
```

```python
cv_ = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=3)
```

```python
# Hyperparameter Tuning by  GridSearchCV
grid={
    'criterion':['entropy'],
    'max_depth':[None,2,4,6,8,10,12,14],
    'min_samples_split':[4,6,10]
}
rrf=RandomForestClassifier()
grid_search = GridSearchCV(estimator=rrf, param_grid=grid, n_jobs=-1, cv=cv_, scoring='accuracy',error_score=0,verbose=10)
grid_result = grid_search.fit(X_train, under_y)
```

```python
print(f'Best score: {grid_result.best_score_:.4f} have parameters {grid_result.best_params_}')
```

```python
rnn=RandomForestClassifier(criterion='entropy',min_samples_split=4)
rnn.fit(X_train,under_y)
train_score=cross_val_score(rnn,X_train, under_y, scoring='accuracy', cv=cv_, n_jobs=-1)
vali_score=cross_val_score(rnn,X_vali, y_vali, scoring='accuracy', cv=cv_, n_jobs=-1)

print(f'Mean accuracy on train data {np.mean(train_score):.4f}')
print(f'Mean accuracy on validation data {np.mean(vali_score):.4f}')
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```
